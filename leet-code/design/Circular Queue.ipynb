### Design Circular Queue
<pre>
Design your implementation of the circular queue. The circular queue is a linear data structure in which the operations are performed based on FIFO (First In First Out) principle and the last position is connected back to the first position to make a circle. It is also called "Ring Buffer".

One of the benefits of the circular queue is that we can make use of the spaces in front of the queue. In a normal queue, once the queue becomes full, we cannot insert the next element even if there is a space in front of the queue. But using the circular queue, we can use the space to store new values.

Your implementation should support following operations:

MyCircularQueue(k): Constructor, set the size of the queue to be k.
Front: Get the front item from the queue. If the queue is empty, return -1.
Rear: Get the last item from the queue. If the queue is empty, return -1.
enQueue(value): Insert an element into the circular queue. Return true if the operation is successful.
deQueue(): Delete an element from the circular queue. Return true if the operation is successful.
isEmpty(): Checks whether the circular queue is empty or not.
isFull(): Checks whether the circular queue is full or not.
</pre>

### Solution 01 - Using Fixed Size Array

```python
class MyCircularQueue1:

    def __init__(self, k: int):
        """
        Initialize your data structure here. Set the size of the queue to be k.
        """
        self.arr = [0] * k
        self.head = -1
        self.count = 0
        self.capacity = k

    def enQueue(self, value: int) -> bool:
        """
        Insert an element into the circular queue. Return true if the operation is successful.
        """
        if not self.isFull():
            self.count += 1
            tail = (self.head + self.count - 1) % self.capacity
            self.arr[tail] = value
            return True
        else:
            return False

    def deQueue(self) -> bool:
        """
        Delete an element from the circular queue. Return true if the operation is successful.
        """
        if not self.isEmpty():
            self.head = (self.head + 1) % self.capacity
            self.count -= 1
            return True
        else:
            return False

    def Front(self) -> int:
        """
        Get the front item from the queue.
        """
        if not self.isEmpty():
            return self.arr[self.head]
        else:
            return -1

    def Rear(self) -> int:
        """
        Get the last item from the queue.
        """
        if not self.isEmpty():
            tail = (self.head + self.count - 1) % self.capacity
            return self.arr[tail]
        else:
            return -1

    def isEmpty(self) -> bool:
        """
        Checks whether the circular queue is empty or not.
        """
        if self.count == 0:
            return True
        else:
            return False

    def isFull(self) -> bool:
        """
        Checks whether the circular queue is full or not.
        """
        if self.count == self.capacity:
            return True
        else:
            return False
```

```python
obj = MyCircularQueue1(3)
param_1 = obj.enQueue(7)
param_2 = obj.deQueue()
param_3 = obj.Front()
param_4 = obj.Rear()
param_5 = obj.isEmpty()
param_6 = obj.isFull()
print(param_1)
print(param_2)
print(param_3)
print(param_4)
print(param_5)
print(param_6)
```

### Solution 02 - Using Linked List
Linked list based implementation is much efficient in terms of memory as don't have to allocate sapce upfront.

```python
class ListNode():
    def __init__(self, val=None):
        self.val = val
        self.next = None
        
class MyCircularQueue2:

    def __init__(self, k: int):
        """
        Initialize your data structure here. Set the size of the queue to be k.
        """
        self.dummy_head = ListNode()
        self.tail = self.dummy_head
        self.capacity = k
        self.count = 0
        

    def enQueue(self, value: int) -> bool:
        """
        Insert an element into the circular queue. Return true if the operation is successful.
        """
        if not self.isFull():
            self.tail.next = ListNode(value)
            self.tail = self.tail.next
            self.count += 1
            return True
        else:
            return False

    def deQueue(self) -> bool:
        """
        Delete an element from the circular queue. Return true if the operation is successful.
        """
        if not self.isEmpty():
            head = self.dummy_head.next
            self.dummy_head.next = head.next
            self.count -= 1
             # update tail when last element is deleted
            if self.count == 0:
                self.tail = self.dummy_head
            return True
        else:
            return False

    def Front(self) -> int:
        """
        Get the front item from the queue.
        """
        if not self.isEmpty():
            return self.dummy_head.next.val
        else:
            return -1
        

    def Rear(self) -> int:
        """
        Get the last item from the queue.
        """
        if not self.isEmpty():
            return self.tail.val
        else:
            return -1
        

    def isEmpty(self) -> bool:
        """
        Checks whether the circular queue is empty or not.
        """
        if self.count == 0:
            return True
        else:
            return False

    def isFull(self) -> bool:
        """
        Checks whether the circular queue is full or not.
        """
        if self.count == self.capacity:
            return True
        else:
            return False
```

```python
obj = MyCircularQueue2(3)
param_1 = obj.enQueue(7)
param_2 = obj.deQueue()
param_3 = obj.Front()
param_4 = obj.Rear()
param_5 = obj.isEmpty()
param_6 = obj.isFull()
print(param_1)
print(param_2)
print(param_3)
print(param_4)
print(param_5)
print(param_6)
```
